## MATPLOTLİB

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.array([20, 5, 10, 55, 20, 25])
y = np.array([65, 25, 10, 75, 85, 45])

plt.plot(x, y, label = "Öylesine", color = "red", linewidth = 2)  # linestyle = "dotted" soluk görüntü verir. 
                                                                  # linestyle = "dashed" çizgili yapar. 
plt.title("Grafik 1")                                             # linestyle = "dashdot" çizgi ve nok. yapar. 
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.legend()

plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.random.randint(0, 100, 10)
y = np.random.randint(0, 100, 10)

plt.plot(x, y, "o--b", label = "Random")

plt.title("Grafik 2")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.legend()

plt.grid(color = "red", linestyle = "--", linewidth = 0.8)
# plt.grid(axis="x", linewidth = 1.0)

plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

a = np.linspace(0, 100, 25)
x, y = plt.subplots(4)

y[0].plot(a, a, color = "red", label = "Doğrusal")
y[0].set_title("Grafik 1")

y[1].plot(a, a ** 2, color = "blue", label = "Karasel")
y[1].set_title("Grafik 2")

y[2].plot(a, a ** 3, color = "yellow", label = "Küpsel")
y[2].set_title("Grafik 3")

y[3].plot(a, a ** 4, color = "green", label = "Dörtgensel")
y[3].set_title("Grafik 4")

plt.suptitle("Grafik")
plt.legend()

plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

a = np.linspace(0, 100, 25)
x, y = plt.subplots(2, 2)

y[0, 0].plot(a, a, color = "red", label = "Doğrusal")
y[0, 0].set_title("Grafik 1")

y[1, 0].plot(a, a ** 2, color = "blue", label = "Karasel")
y[1, 0].set_title("Grafik 2")

y[0, 1].plot(a, a ** 3, color = "yellow", label = "Küpsel")
y[0, 1].set_title("Grafik 3")

y[1, 1].plot(a, a ** 4, color = "green", label = "Dörtgensel")
y[1, 1].set_title("Grafik 4")

plt.suptitle("Grafik")
plt.legend()

plt.show()
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

x = pd.read_csv("weatherAUS.csv")
y = x.groupby("Temp3pm").mean()
y.plot(subplots = True)

plt.legend()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(0, 10, 50)
y = x ** 2
z = x ** 3

fig = plt.figure()

q = fig.add_axes([1.2, 1.2, 1.2, 1.2])    # soldan, altdan, sağdan, üstten

q.plot(x, y, "b--o", label = "Karasel")
q.plot(x, z, "r--o", label = "Küpsel")

q.set_xlabel("X Ekseni")
q.set_ylabel("Y Ekseni")

plt.legend(loc = 2)   # 1 >> sağ üst, 2 >> sol üst, 3 >> sol alt, 4 >> sağ alt
plt.grid()
plt.title("GRAFİK")
plt.show()
```

```python
import numpy as np              
import matplotlib.pyplot as plt

month = ["Ocak", "Şubart", "Mart", "Nisan", "Mayıs", "Haziran", "Temmuz", "Ağustos", "Eylül", "Ekim", "Kasım", "Aralık"]

audio = [3000, 2000, 1500, 5000, 1200, 2400, 1300, 2600, 2500, 500, 700, 300]
renault = [200, 3000, 4000, 1230, 3100, 140, 500, 600, 2400, 200, 1300, 4200]
BMW = [100, 200, 400, 300, 500, 600, 700, 1000, 1200, 1400, 1500, 1600]

plt.plot([],[], color = "yellow", label = "Audio")
plt.plot([],[], color = "blue", label = "Renault")
plt.plot([],[], color = "red", label = "BMW")

plt.stackplot(month, audio, renault, BMW, colors=["yellow", "blue", "red"])

plt.legend(loc = 2)

plt.xlabel("Aylar")
plt.ylabel("Araba satış sayıları")

plt.title("Aylara göre araba satışları")
plt.show()
```

## Pasta Grafiği

```python
import matplotlib.pyplot as plt 

fruits = ["Apple", "Orange", "Banana", "Date"]
color = ["green", "orange", "yellow", "red"]
ton_year = [25000, 6500, 18000, 30000]

plt.pie(ton_year, labels=fruits, colors = color, shadow = True, explode = (0.1, 0.1, 0.1, 0.1), autopct ="%1.2f%%")

# plt.legend()

plt.show()
```
